```python
#Dashboard
relevance_criteria = {}
num_criteria = int(input("Enter the number of relevance criteria: "))
for _ in range(num_criteria):
    relevance_level = input("Enter relevance level (e.g., High, Medium, Low): ")
    keywords = input("Enter keywords separated by commas (e.g., tutorial, guide, step-by-step): ").split(",")
    relevance_criteria[relevance_level] = [keyword.strip() for keyword in keywords]

# Prompt the user to enter suggestions
num_suggestions = int(input("Enter the number of suggestions: "))
suggestions = []
for i in range(num_suggestions):
    suggestion = input(f"Enter suggestion {i+1}: ")
    suggestions.append(suggestion)

# Customizing the display of suggestions based on relevance criteria
print("=================================")
print("         Auto-suggested Queries        ")
print("=================================")
for index, suggestion in enumerate(suggestions, start=1):
    relevance = "Low"
    for level, keywords in relevance_criteria.items():
        if any(keyword in suggestion.lower() for keyword in keywords):
            relevance = level
            break
    print(f"{index}. [{relevance}] {suggestion}")
print("=================================")
```

```python
#Q2
#a
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
Input = pd.read_csv("diabetes.csv")

# Set seed for reproducibility
np.random.seed(746)

# Take a random sample of 25 observations
sample_data = Input.sample(n=25)

# Calculate mean and highest Glucose values for both sample and population
mean_of_the_sample_glucose = sample_data['Glucose'].mean()
highest__of_the_sample_glucose = sample_data['Glucose'].max()

mean_population_glucose = Input['Glucose'].mean()
mean_highest_glucose = Input['Glucose'].max()

# Create a bar chart to compare mean Glucose values
plt.bar(['Sample', 'Population'], [mean_of_the_sample_glucose, mean_population_glucose], color=['black', 'yellow'])
plt.title('Mean Glucose Comparison')
plt.xlabel('Dataset')
plt.ylabel('Mean Glucose')
plt.show()

# Create a bar chart to compare highest Glucose values
plt.bar(['Sample', 'Population'], [highest__of_the_sample_glucose, mean_highest_glucose], color=['black', 'yellow'])
plt.title('Highest Glucose Comparison')
plt.xlabel('Dataset')
plt.ylabel('Highest Glucose')
plt.show()
```
